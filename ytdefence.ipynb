```python
import os
os.makedirs("yt_processed", exist_ok=True)

import numpy as np

def chunk_audio(input_path, base_name, chunk_length=10):

    audio, sr = librosa.load(input_path, sr=16000)

    duration = len(audio) / sr
    print(f"{base_name} duration: {round(duration,2)} sec")

    chunk_samples = chunk_length * sr
    total_chunks = len(audio) // chunk_samples

    print(f"Chunks to create: {total_chunks}")

    paths = []

    for i in range(total_chunks):

        chunk = audio[i*chunk_samples:(i+1)*chunk_samples]

        out_path = f"yt_processed/{base_name}_chunk_{i}.wav"

        sf.write(out_path, chunk, sr)

        paths.append(out_path)

    return paths
```

```python
all_chunks = []

for file in os.listdir("yt_processed"):

    input_path = os.path.join("yt_processed", file)
    base_name = file.split(".")[0]

    chunks = chunk_audio(input_path, base_name, chunk_length=5)

    all_chunks.extend(chunks)

print("Total chunks:", len(all_chunks))
```

```python
def trim_audio(path):

    audio, sr = librosa.load(path, sr=16000)

    trimmed, _ = librosa.effects.trim(audio)

    sf.write(path, trimmed, sr)

for p in all_chunks:
    trim_audio(p)
```

```python
import re

def clean_text(text):

    text = text.lower()

    # remove common youtube phrases
    remove_phrases = [
        "thank you for watching",
        "subscribe",
        "like and share",
        "click the bell icon",
        "see you in the next video"
    ]

    for p in remove_phrases:
        text = text.replace(p, "")

    # remove special characters
    text = re.sub(r"[^\w\s]", "", text)

    # normalize spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
import os
import librosa

def check_dataset_hours(directory):
    total_seconds = 0
    audio_extensions = ('.wav', '.mp3', '.flac', '.m4a')

    # Get all audio files from your combined directory
    audio_files = [f for f in os.listdir(directory) if f.endswith(audio_extensions)]

    if not audio_files:
        print("No audio files found in the directory.")
        return

    print(f"Analyzing {len(audio_files)} files in {directory}...")

    for filename in audio_files:
        file_path = os.path.join(directory, filename)
        try:
            # Load metadata only to save time (doesn't load full audio)
            duration = librosa.get_duration(path=file_path)
            total_seconds += duration
        except Exception as e:
            print(f"Error processing {filename}: {e}")

    total_hours = total_seconds / 3600

    print("-" * 30)
    print(f"Total Duration: {total_hours:.2f} hours")
    print(f"Total Files: {len(audio_files)}")
    print("-" * 30)

# Path to your combined folder
dataset_path = '/content/drive/MyDrive/final_dataset/audio'
check_dataset_hours(dataset_path)
```

```python
import os
import librosa
import soundfile as sf
import numpy as np

# Paths from your Week 3 setup
input_dir = '/content/drive/MyDrive/final_dataset/processed_defence'
output_dir = '/content/drive/MyDrive/final_dataset/hifi_defence'
os.makedirs(output_dir, exist_ok=True)

def estimate_snr_simple(audio):
    # Calculate Root Mean Square (RMS) energy
    rms = np.sqrt(np.mean(audio**2))
    # Estimate noise floor from the quietest 10% of the audio
    sorted_audio = np.sort(np.abs(audio))
    noise_floor = np.mean(sorted_audio[:int(len(sorted_audio)*0.1)])

    if noise_floor == 0: return 50 # Clean signal

    snr = 20 * np.log10(rms / (noise_floor + 1e-10))
    return snr

print("Restarting Step 1: Adjusted Hi-Fi Filtering...")

processed_count = 0
for file_name in os.listdir(input_dir):
    if file_name.endswith('.wav'):
        path = os.path.join(input_dir, file_name)
        audio, sr = librosa.load(path, sr=16000)

        snr = estimate_snr_simple(audio)

        # Lowering threshold slightly to 15dB to ensure we get some data
        # while still aiming for the 'High SNR' requirement
        if snr > 15:
            # Apply LUFS-like normalization (Peak to -1.0)
            norm_audio = audio / (np.max(np.abs(audio)) + 1e-10)

            save_path = os.path.join(output_dir, f"hifi_{file_name}")
            sf.write(save_path, norm_audio, 16000)
            processed_count += 1
            print(f"✅ Accepted: {file_name} (SNR: {snr:.2f}dB)")
        else:
            print(f"❌ Still Skipped: {file_name} (SNR: {snr:.2f}dB)")

print(f"\nTotal files successfully prepared for Identity Locking: {processed_count}")
```

```python
# Install requirement for speaker embeddings
!pip install speechbrain

from speechbrain.pretrained import EncoderClassifier
import torch
import os
import json
import librosa

# Paths
hifi_dir = '/content/drive/MyDrive/final_dataset/hifi_defence'
embedding_output = '/content/drive/MyDrive/final_dataset/speaker_identities.json'

# Load the ECAPA-TDNN model
classifier = EncoderClassifier.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb")

print("Starting Step 2: Extracting Speaker Embeddings (Identity Locking)...")

embeddings = {}
hifi_files = [f for f in os.listdir(hifi_dir) if f.endswith('.wav')]

for file_name in hifi_files:
    path = os.path.join(hifi_dir, file_name)
    try:
        signal, fs = librosa.load(path, sr=16000)

        # Extract identity embedding [Requirement: Intern 1 & 2]
        with torch.no_grad():
            # encode_batch returns a tensor on GPU
            emb = classifier.encode_batch(torch.tensor(signal).unsqueeze(0))

            # .cpu() moves it to host memory, .numpy() converts it
            embeddings[file_name] = emb.squeeze().cpu().numpy().tolist()

        if len(embeddings) % 10 == 0:
            print(f"Processed {len(embeddings)}/{len(hifi_files)} identities...")

    except Exception as e:
        print(f"Error processing {file_name}: {e}")

# Save the 'Identity Lock' file
with open(embedding_output, 'w') as f:
    json.dump(embeddings, f)

print(f"✅ Success: Speaker identities locked in {embedding_output}")
```

```python
import json
import re
import os

# 1. Setup paths
data_dir = '/content/drive/MyDrive/Defence_Phase1_Handover'
vocab_path = os.path.join(data_dir, 'tactical_vocab.json')
parallel_path = os.path.join(data_dir, 'defence_parallel.json')

# 2. Define the corrected and expanded vocabulary
# These additions fix the "Visual on point" and "Oscar-Mike" mismatches
updated_vocab = {
    "Alpha": "Lead Element",
    "Bravo": "Support Element",
    "Copy": "Confirmed",
    "Requesting exfil": "requesting extraction",
    "Visual on point": "Sight contact on",
    "Oscar-Mike": "on the move",
    "point Bravo": "Support Element"
}

# 3. Write the updated vocabulary back to the file
try:
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)

    with open(vocab_path, 'w') as f:
        json.dump(updated_vocab, f, indent=4)
    print(f"✅ Successfully updated {vocab_path}")
except Exception as e:
    print(f"❌ Failed to update vocabulary: {e}")

# 4. Function to apply the mapping (including normalization)
def transform_audio_transcript(source_text, mapping):
    # Remove leading commas and whitespace often found in raw audio transcripts
    clean_text = re.sub(r'^[\s,.]+', '', source_text).strip()

    # Sort keys by length (longest first) to ensure "Visual on point"
    # is replaced before the individual word "point"
    sorted_keys = sorted(mapping.keys(), key=len, reverse=True)

    transformed_text = clean_text
    for key in sorted_keys:
        # Use word boundaries (\b) to avoid replacing parts of other words
        pattern = re.compile(rf'\b{re.escape(key)}\b', flags=re.IGNORECASE)
        transformed_text = pattern.sub(mapping[key], transformed_text)

    # Ensure the sentence ends with a period to match target formatting
    if not transformed_text.endswith('.'):
        transformed_text += '.'

    return transformed_text

# 5. Verify the fix
if os.path.exists(parallel_path):
    with open(parallel_path, 'r') as f:
        parallel_data = json.load(f)

    print("\n--- Verification Audit ---")
    for entry in parallel_data:
        result = transform_audio_transcript(entry['source'], updated_vocab)
        print(f"Source: {entry['source']}")
        print(f"Result: {result}")

        # Strip trailing whitespace/periods for a fair comparison
        if result.lower().strip('.') == entry['target'].lower().strip('.'):
            print("✅ FIXED: Match found.")
        else:
            print(f"❌ STILL MISMATCHED: Expected '{entry['target']}'")
```

```python
import json
import os
import re
import wave

# Configuration
base_dir = '/content/drive/MyDrive/Defence_Phase1_Handover'
audio_folder = os.path.join(base_dir, '/content/drive/MyDrive/Defence_Phase1_Handover/hifi_defence')
metadata_path = os.path.join(base_dir, 'aligned_metadata.json')

def get_duration(file_path):
    """Returns the duration of a wav file in seconds."""
    try:
        with wave.open(file_path, 'r') as f:
            frames = f.getnframes()
            rate = f.getframerate()
            return frames / float(rate)
    except Exception:
        return 0

def generate_aligned_metadata():
    if not os.path.exists(audio_folder):
        print(f"❌ Error: Folder {audio_folder} not found.")
        return

    metadata_entries = []

    # 1. Get and Sort Files Numerically
    audio_files = [f for f in os.listdir(audio_folder) if f.endswith('.wav')]
    audio_files.sort(key=lambda f: int(re.sub('\D', '', f)) if re.sub('\D', '', f) else 0)

    elapsed_non_yt_seconds = 0
    ATC_LIMIT_SECONDS = 3600  # 1 Hour

    for filename in audio_files:
        full_path = os.path.join(audio_folder, filename)
        duration = get_duration(full_path)
        fn_lower = filename.lower()

        # LOGIC GATE 1: YouTube (Tactical)
        if "yt" in fn_lower:
            corpus = "YouTube_Field_Recordings"
            speaker_id = "Unit_Lead_Alpha"
            quality = "Noisy, stressed, tactical jargon"

        # LOGIC GATE 2: ATC (First hour of non-YT)
        elif elapsed_non_yt_seconds < ATC_LIMIT_SECONDS:
            corpus = "ATC_Tactical_Radio"
            speaker_id = "Tower_Control"
            quality = "Static, rhythmic, high-speed"
            elapsed_non_yt_seconds += duration

        # LOGIC GATE 3: TIMIT (Everything after the first hour of non-YT)
        else:
            corpus = "TIMIT_Phonetic_Base"
            speaker_id = "Neutral_Speaker"
            quality = "Studio quality, perfect pronunciation"

        metadata_entries.append({
            "audio_file": filename,
            "corpus": corpus,
            "speaker_id": speaker_id,
            "expected_quality": quality,
            "duration_sec": round(duration, 2)
        })

    # Write to JSON
    with open(metadata_path, 'w') as f:
        json.dump(metadata_entries, f, indent=4)

    print(f"✅ Metadata Complete.")
    print(f"Total files processed: {len(metadata_entries)}")
    print(f"ATC Data identified: {round(elapsed_non_yt_seconds/60, 2)} minutes.")

if __name__ == "__main__":
    generate_aligned_metadata()
```

```python
import json
import os
import re
import wave

# Configuration
base_dir = '/content/drive/MyDrive/Defence_Phase1_Handover'
audio_folder = os.path.join(base_dir, '/content/drive/MyDrive/Defence_Phase1_Handover/hifi_defence')
metadata_path = os.path.join(base_dir, 'aligned_metadata.json')

def get_duration(file_path):
    try:
        with wave.open(file_path, 'r') as f:
            return f.getnframes() / float(f.getframerate())
    except Exception:
        return 0

def extract_number(filename):
    """Extracts the last number in a filename for proper sorting."""
    nums = re.findall(r'\d+', filename)
    return int(nums[-1]) if nums else 0

def generate_aligned_metadata():
    if not os.path.exists(audio_folder):
        print(f"❌ Error: Folder {audio_folder} not found.")
        return

    # 1. Separate YouTube and Non-YouTube files
    all_files = [f for f in os.listdir(audio_folder) if f.endswith('.wav')]
    yt_files = [f for f in all_files if "yt" in f.lower()]
    non_yt_files = [f for f in all_files if "yt" not in f.lower()]

    # 2. Sort Non-YT files strictly by the number (0, 1, 2, 3...)
    non_yt_files.sort(key=extract_number)

    metadata_entries = []
    elapsed_atc_seconds = 0
    ATC_LIMIT_SECONDS = 3600  # 1 Hour exact

    # PROCESS NON-YOUTUBE (ATC then TIMIT)
    for filename in non_yt_files:
        full_path = os.path.join(audio_folder, filename)
        duration = get_duration(full_path)

        if elapsed_atc_seconds < ATC_LIMIT_SECONDS:
            corpus = "ATC_Tactical_Radio"
            speaker_id = "Tower_Control"
            quality = "Static, rhythmic, high-speed"
            elapsed_atc_seconds += duration
        else:
            corpus = "TIMIT_Phonetic_Base"
            speaker_id = "Neutral_Speaker"
            quality = "Studio quality, perfect pronunciation"

        metadata_entries.append({
            "audio_file": filename,
            "corpus": corpus,
            "speaker_id": speaker_id,
            "expected_quality": quality,
            "duration_sec": round(duration, 2)
        })

    # PROCESS YOUTUBE (Independent of the 1-hour clock)
    for filename in yt_files:
        duration = get_duration(os.path.join(audio_folder, filename))
        metadata_entries.append({
            "audio_file": filename,
            "corpus": "YouTube_Field_Recordings",
            "speaker_id": "Unit_Lead_Alpha",
            "expected_quality": "Noisy, stressed, tactical jargon",
            "duration_sec": round(duration, 2)
        })

    with open(metadata_path, 'w') as f:
        json.dump(metadata_entries, f, indent=4)

    print(f"✅ Metadata Re-aligned.")
    print(f"ATC Duration: {round(elapsed_atc_seconds/60, 2)} minutes")
    print(f"Total Metadata Entries: {len(metadata_entries)}")

if __name__ == "__main__":
    generate_aligned_metadata()
```

```python
import json
import os

# Configuration
base_dir = '/content/drive/MyDrive/Defence_Phase1_Handover'
metadata_path = os.path.join(base_dir, 'aligned_metadata.json')
parallel_path = os.path.join(base_dir, 'defence_parallel.json')
manifest_path = os.path.join(base_dir, 'final_train_manifest.json')

def create_final_manifest():
    if not os.path.exists(metadata_path) or not os.path.exists(parallel_path):
        print("❌ Error: Missing required JSON files in the handover directory.")
        return

    with open(metadata_path, 'r') as f:
        metadata = json.load(f)
    with open(parallel_path, 'r') as f:
        parallel_data = json.load(f)

    final_manifest = []

    for meta, text in zip(metadata, parallel_data):
        filename = meta['audio_file']

        # Globally replace 'audio/' with 'hifi_defence/'
        # to ensure the folder structure is correct.
        corrected_path = os.path.join("hifi_defence", filename)

        entry = {
            "id": filename.replace('.wav', ''),
            "audio_path": corrected_path, # Result: hifi_defence/hifi_prep_audio_0.wav
            "corpus": meta['corpus'],
            "speaker_id": meta['speaker_id'],
            "source_text": text['source'],
            "target_text": text['target'],
            "duration_sec": meta.get('duration_sec', 0)
        }
        final_manifest.append(entry)

    with open(manifest_path, 'w') as f:
        json.dump(final_manifest, f, indent=4)

    print(f"🚀 Final Manifest Synchronized.")
    print(f"Verified Path Sample: {final_manifest[0]['audio_path']}")

if __name__ == "__main__":
    create_final_manifest()
```

```python
import os
import wave
import datetime

# --- CONFIGURATION ---
audio_folder = '/content/drive/MyDrive/Defence_Phase1_Handover/hifi_defence' # Update this to your folder name
# ---------------------

def get_folder_duration(folder_path):
    total_seconds = 0
    file_count = 0

    if not os.path.exists(folder_path):
        print(f"❌ Error: Folder '{folder_path}' not found.")
        return

    for filename in os.listdir(folder_path):
        if filename.endswith('.wav'):
            file_path = os.path.join(folder_path, filename)
            try:
                with wave.open(file_path, 'r') as f:
                    frames = f.getnframes()
                    rate = f.getframerate()
                    duration = frames / float(rate)
                    total_seconds += duration
                    file_count += 1
            except Exception as e:
                print(f"⚠️ Could not read {filename}: {e}")

    # Convert seconds to a readable H:M:S format
    readable_time = str(datetime.timedelta(seconds=int(total_seconds)))
    total_hours = total_seconds / 3600

    print(f"--- Audio Audit for '{folder_path}' ---")
    print(f"Total Files:  {file_count}")
    print(f"Total Time:   {readable_time} (HH:MM:SS)")
    print(f"Total Hours:  {round(total_hours, 2)} hours")

if __name__ == "__main__":
    get_folder_duration(audio_folder)
```
